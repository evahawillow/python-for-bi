#Task1: list of dictionaries

```python
transactions = [
    {"id": 1, "product": "Laptop", "price": 1200, "quantity": 2, "region": "EU"},
    {"id": 2, "product": "Phone", "price": 800, "quantity": -1, "region": "US"},
    {"id": 3, "product": "Tablet", "price": 0, "quantity": 5, "region": "EU"},
    {"id": 4, "product": "Monitor", "price": 350, "quantity": 3, "region": "US"},
    {"id": 5, "product": "Laptop", "price": 1200, "quantity": 1, "region": "EU"},
    {"id": 6, "product": "Keyboard", "price": 75, "quantity": 10, "region": "APAC"},
]

total_revenue_by_region = {}

for i in transactions:
    #a. filtering invalid transactions
    if i["price"] > 0 and i["quantity"] >= 0:
        # b) compute total
        i["total"] = i["price"] * i["quantity"]
        # c) group by region
        region = i["region"]
        if region not in total_revenue_by_region:
            total_revenue_by_region[region] = i["total"]
        else:
            total_revenue_by_region[region] += i["total"]  

#d) Print the region with the highest total revenue            
print(f"The region with the highest total revenue is {max(total_revenue_by_region.values())}€")    
```

#Task2: function

```python
sales=[1200, 980, 1540, 870, 2010, 1650, 1300, 1100, 950, 1780, 2200, 1430, 1600, 1900]

def summarise_sales(sales):
    #"total": the sum of all daily sales
    total = 0
    for sale in sales:
        total = total+sale
   
    #"average": the mean daily sales value, rounded to two decimals
    average = round(total / len(sales), 2)
    
    #"best_day": the index (starting from 1) of the day with the highest sales
    best_day = sales.index(max(sales)) + 1  # +1 to start counting days from 1
    
    #"worst_day": the index (starting from 1) of the day with the lowest sales
    worst_day = sales.index(min(sales)) + 1
 
    
    #"above_average_count": the number of days where sales exceeded the average
    above_average_count=0
    for sale in sales:
        if sale>average:
            above_average_count +=1
    
    #returning dictionary
    return {
        "total": total,
        "average": average,
        "best_day": best_day,
        "worst_day": worst_day,
        "above_average_count": above_average_count
    }

#calling a function and storing it as a "summary"
summary = summarise_sales(sales)

#printing the dictionary
print(summary)
```

#Task3: Loops

```python
students = [ 
    {"name": "Anna", "grade": 85, "attendance": 0.92}, 
    {"name": "Erik", "grade": 58, "attendance": 0.75}, 
    {"name": "Mia", "grade": 91, "attendance": 0.98}, 
    {"name": "Leo", "grade": 44, "attendance": 0.60}, 
    {"name": "Sara", "grade": 72, "attendance": 0.88}, 
    {"name": "Tom", "grade": 67, "attendance": 0.70}, 
    {"name": "Nina", "grade": 95, "attendance": 0.99}, 
    {"name": "Olli", "grade": 50, "attendance": 0.65}, 
] 

#a
"""Creating the new passing list with eligible name and the grade above 50"""

passing = [] 
for i in students: 
    if i["grade"]>=50: 
        passing.append(i["name"])

print(f"The students passing the course are {passing}")


#b 
"""Looping through students and printing the message"""
for i in students:
    if i["grade"]>=80:
        print(f"\nStudent {i["name"]}: Excellent")
    elif i["grade"]>=50 and i["grade"]<=79:
        print(f"Student {i["name"]}: Pass")
    else:
        print(f"Student {i["name"]}: Fail")

#c
"""Counting students with attendance rate below 0.80 and printing"""
count_low_attendance=0

for i in students:
    if i["attendance"]<0.8:
        count_low_attendance +=1
print(f"\nThe number of students with the attendance rate below 0.8 is {count_low_attendance}")

#d 
"""Finding the student with the highest grade and printing"""

best_student=students[0]

for i in students:
    if i["grade"]>best_student["grade"]:
        best_student=i

print(f"\nBest student is {best_student["name"]} with the grade {best_student["grade"]}")

#e
"""Finding the student with the highest grade using max() and printing"""
max_grade = max([i["grade"] for i in students]) #finding the highest grade

for student in students: # finding the student with that grade
    if student["grade"] == max_grade:
        best_student = student
        break

print(f"\nBest student is {best_student['name']} with the grade {best_student['grade']}")
```

#Task4: Class and methods

```python
"""Design a class called Portfolio that represents an investment portfolio"""

class Portfolio:
    def __init__(self, owner):
        self.owner = owner
        self.holdings = {} 
    
    #a) A method add_asset(ticker, shares, price_per_share) that adds a new asset or updates the number of shares if the ticker already exists (in which case the price should also be updated).
    def add_asset(self, ticker, shares, price_per_share):
        if ticker in self.holdings:
        #updating existing asset
            self.holdings[ticker]["shares"] += shares
            self.holdings[ticker]["price_per_share"] = price_per_share
        else:
        #adding new asset
            self.holdings[ticker] = {
                "shares": shares,
                "price_per_share": price_per_share
            }
    
    #b)  A method total_value calculating and returning total value of the portfolio   
    def total_value(self):
        total=0
        for asset in self.holdings.values():
            total += asset["shares"]*asset["price_per_share"]
        return total
    
    #c) A method most_valuable() that returns the ticker of the asset with the highest total value (shares × price).
    def most_valuable(self):
        #setting empty values
        max_value = 0
        best_ticker = None

        #looping through dictionary, getting access to the asets and calculating the highest total value
        for ticker in self.holdings:
            asset = self.holdings[ticker]
            value = asset["shares"]*asset["price_per_share"]
            
            if value > max_value:
                max_value = value
                best_ticker = ticker

        return best_ticker
                
    #d) A __str__ method that returns a readable summary of the portfolio, including the owner's name, the number of assets, and the total value.
    def __str__(self):
        return(f"\nInvestor Portfolio of {self.owner} contains {len(self.holdings)} assets and its total value amounts for {self.total_value():,.0f}€")

#Creating an instance, adding several assets (Wärstilä, Nordea, Finnair, Neste, Nokia), and demonstrating all methods with printed output.

#creating an object/instance
investor_portfolio = Portfolio("Alexandra")

#adding assets
investor_portfolio.add_asset("WRT1V", 20, 31.37)
investor_portfolio.add_asset("NDA-FI", 50, 14.29)
investor_portfolio.add_asset("FIA1S", 30, 2.91)
investor_portfolio.add_asset("NESTE", 40, 26.93)
investor_portfolio.add_asset("NOK", 25, 6.87)

#total value
print(f"\nTotal value of Investor Portfolio is {investor_portfolio.total_value()} €")

#most valuable asset
print(f"\nMost valuable asset is {investor_portfolio.most_valuable()}")

#printing portfolio summary 
print(investor_portfolio)

#updating portfolio by adding 15 shares to Nokia and updating price 
investor_portfolio.add_asset("NOK", 40, 5.50)

# Print updated portfolio
print(f"\nAfter updating NOK the {investor_portfolio}")

print(f"\nTotal value of Investor Portfolio is {investor_portfolio.total_value()} €")
print(f"\nMost valuable asset is {investor_portfolio.most_valuable()}")

    
```

#Task5: Importing a module

```python
#importing datatime module
import datetime

orders = [
    ("2025-01-15", "Laptop"), ("2025-01-22", "Phone"),
    ("2025-02-03", "Laptop"), ("2025-02-14", "Tablet"),
    ("2025-03-01", "Phone"), ("2025-03-10", "Phone"),
    ("2025-03-22", "Laptop"), ("2025-04-05", "Monitor"),
]


dates = [] #creating empty list to store the converted date objects

for order in orders: #looping through each order
    date_obj = datetime.date.fromisoformat(order[0])  #converting date strings to datetime.date objects using fromisoformat() as data strings are in ISO format (YYYY-MM-DD)
    dates.append(date_obj) #adding new date object to the dates list

#finding earliest and latest date
earliest = min(dates)
latest = max(dates)

#computing number of days between them
days_between = (latest - earliest).days

#printing results
print(f"Earliest order is on {earliest}")
print(f"Latest order is on {latest}")
print(f"Number of days between earliest and latest order is {days_between}")
```

#Task 6: Array Operations and Broadcasting 

```python
import numpy as np

# Revenue data: rows = years (2022, 2023, 2024), columns = quarters (Q1, Q2, Q3, Q4)
revenue = np.array([
    [210, 230, 190, 250],
    [220, 260, 200, 270],
    [250, 290, 230, 310]
])

#a)computing and printing total annual revenue using axis parameter (across rows).
total_annual = revenue.sum(axis=1)
print(f"Total annual revenue is {total_annual}")

#b)computing mean revenue per quarter (mean across columns)
mean_quarterly = revenue.mean(axis=0)
print(f"Mean quarterly revenue is {mean_quarterly}")

#c)adjusting growth to 2024 [2]=250,290, 230, 210 multiplying the year quartley revenues*growth array values
growth = np.array([1.05, 1.03, 1.08, 1.02])
adjusted_2024 = revenue[2] * growth
print(f"Adjusted 2024 revenue is {adjusted_2024}")

#d)finiding and printing quarterly revenue values above 250 
high_revenues = revenue[revenue > 250]
print(f"Revenue values above 250 are {high_revenues}")
```

#Task 7: Pandas

```python
import pandas as pd

#creating the DataFrame

data = {
    "employee": ["Anna", "Erik", "Mia", "Leo", "Sara", "Tom"],
    "department": ["Finance", "Marketing", "Finance", "IT", "Marketing", "IT"],
    "salary": [4500, 4000, 4800, 5200, 3900, 5100],
    "years_exp": [5, 3, 7, 10, 2, 8],
    "performance_score": [4.2, 3.8, 4.5, 4.0, 3.5, 4.9]
}

employees = pd.DataFrame(data)

#a)adding a new column salary_per_year_exp computed as salary / years_exp + rounding

employees["salary_per_year_exp"] = (employees["salary"]/employees["years_exp"]).round()
print(employees)

#b)sorting the DataFrame by performance_score in descending order, by using method .sort_values, ascending=False (descendinding order) parameter and head.() method to get the 3 top rows.

top_employees = employees.sort_values(by="performance_score", ascending = False).head(3)
print(top_employees)

#c)selecting all employees in the Finance department with a performance score above 4.0, using .loc

finance_employees = employees.loc[
    (employees["department"] == "Finance") & (employees ["performance_score"] > 4.0)
     ]
print(finance_employees)

#d)renaming the column years_exp to experience_years using the .rename() method

employees = employees.rename(columns={"years_exp": "experience_years"})
print(employees)

#e)Selecting salary and performance score columns for the first four employees, using .iloc.
four_employees = employees.iloc[0:4, [2, 4]]
print(four_employees)
```

#Task8: More Pandas

```python
#A company records daily website traffic over four weeks. Generate the data by running the following script first:

import numpy as np
import pandas as pd

#generating the data

rng = np.random.default_rng(seed=123)
days = pd.date_range(start="2025-01-06", periods=28, freq="D")
visitors = rng.integers(200, 600, size=28)
conversion_rate = rng.uniform(0.02, 0.08, size=28).round(4)
avg_order_value = rng.normal(50, 10, size=28).round(2)

#a)creating a DataFrame with columns: date, visitors, conversion_rate, and avg_order_value
traffic = pd.DataFrame({
    #"date": days,
    "visitors": visitors,
    "conversion_rate": conversion_rate,
    "avg_order_value": avg_order_value
}, index = days)

#setting the date as index
#traffic.set_index("date", inplace=True)

#b)computing two new columns: conversions (= visitors × conversion_rate, rounded to the nearest integer) and daily_revenue (= conversions × avg_order_value)
traffic["conversions"] = (traffic["visitors"]/traffic["conversion_rate"]).round()
traffic["daily_revenue"] = (traffic["conversions"] * traffic["avg_order_value"]).round()

print(traffic)
```
